## <font color='blue'>Espaço Vetorial $\mathbb{R}^n$</font>

Um __vetor__ $n$-dimensional $\mathbf{x}$ é uma matriz com $n$ linhas e uma coluna. 

$$
\mathbf{x}=\begin{bmatrix}
x_1 \\ x_2 \\ \vdots \\ x_n
\end{bmatrix}
$$

Vetores podem ser somados e multiplicados por um escalar, dando origem a outros vetores.

$$
\mathbf{x}=\begin{bmatrix}
x_1 \\ x_2 \\ \vdots \\ x_n
\end{bmatrix},\,
\mathbf{y}=\begin{bmatrix}
y_1 \\ y_2 \\ \vdots \\ y_n
\end{bmatrix}\quad \Longrightarrow\quad
\mathbf{x}+\mathbf{y}=\begin{bmatrix}
x_1+y_1 \\ x_2+y_2 \\ \vdots \\ x_n+y_n
\end{bmatrix},\quad
c\mathbf{x}=\begin{bmatrix}
cx_1 \\ cx_2 \\ \vdots \\ cx_n
\end{bmatrix}
$$

O espaço $\mathbb{R}^n$ é o espaço que contém todos os vetores reais $n$-dimensionais.

Na verdade, $\mathbb{R}^n$ é um __espaço vetorial__. O que caracteriza um espaço vetorial é o fato dos vetores (elementos) de $\mathbb{R}^n$  satisfazerem as seguintes propriedades:

1. $\mathbf{x} + \mathbf{y} = \mathbf{y} + \mathbf{x}$ (simetria)
2. $(\mathbf{x} + \mathbf{y}) + \mathbf{z} = \mathbf{x} + (\mathbf{y} + \mathbf{z})$ (associativa)
3. $\mathbf{x} + \mathbf{0} = \mathbf{x}$ (existe um elemento nulo)
4. $\mathbf{x} + (\mathbf{-x}) = \mathbf{x} - \mathbf{x} = \mathbf{0}$ (existe um elemento inverso)
5. $c(\mathbf{x} + \mathbf{y}) = c\mathbf{x} + c\mathbf{y}$ (distributiva)
6. $(c+d)\mathbf{x} = c\mathbf{x}+d\mathbf{x}$
7. $c(d\mathbf{x}) = (cd)\mathbf{x}$
8. $1\mathbf{x} = \mathbf{x}$ (existe um elemento neutro)

Nas propriedades acima, os elementos $\mathbf{x}, \mathbf{y}, \mathbf{z}$ e $\mathbf{0}$ (vetor nulo) são vetores em $\mathbb{R}^n$ e $c,e$ e $1$ são numeros reais (escalares).

Qualquer conjunto onde as operações de adição e multiplicação por escalar estejam definidas e que, quando aplicadas aos elementos do conjunto, tais operações satisfaçam as 8 propriedades acima, corresponde a um espaço vetorial.

Uma operação muito importante entre vetores de $\mathbb{R}^n$ é o **produto escalar**. Dados dois vetores $\mathbf{x}=[x_1,\ldots,x_n$ e $\mathbf{y}=y_1,\ldots,y_n$, o produto escalar entre $\mathbf{x}$ e $\mathbf{y}$, denotado por $<\mathbf{x},\mathbf{y}>$ é definido por:

$$
<\mathbf{x},\mathbf{y}>=x_1y_1+x_2y_2+\cdots+x_ny_n=\sum_i x_iy_i
$$

Note que o produto escalar entre dois vetores é um escalar (não um vetor).

O produto escalar possui muitas propriedades (tais propriedades podem ser encontradas [aqui](https://math.libretexts.org/Bookshelves/Linear_Algebra/Book%3A_A_First_Course_in_Linear_Algebra_(Kuttler)/04%3A_R/4.07%3A_The_Dot_Product)). Em particular:

1. o produto escalar entre dois vetores ortogonais (que formam um ângulo de $90^o$) é sempre igual a zero. Ou seja, se $\mathbf{x},\mathbf{y}$ são ortogonais então $<\mathbf{x},\mathbf{y}>=0$.

2. a norma de um vetor $\mathbf{x}$, denotado por $\|\mathbf{x}\|$, é dados por:
$
\|\mathbf{x}\|=\sqrt{<\mathbf{x},\mathbf{x}>}
$

Dados dois vetores $\mathbf{x}$ e $\mathbf{y}$, a projeção ortogonal de $\mathbf{x}$ em $\mathbf{y}$ é dada por:

$$
proj_{\mathbf{y}}(\mathbf{x}) = <\mathbf{x},\mathbf{y}>\frac{\mathbf{y}}{\|\mathbf{y}\|^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# definindo os vetores x,y,z em R^2
x = np.array([1,1])
y = np.array([-1,1])
z = np.array([0.1,0.7])

# plotando os eixos de R^2
plt.axes().set_aspect('equal')
plt.plot([-1,1],[0,0],color='k')
plt.plot([0,0],[0,1],color='k')

# plotando os vetores
plt.arrow(0,0,x[0],x[1],head_width=0.1,length_includes_head=True)
plt.text(x[0],x[1],'$\mathbf{x}$')
plt.arrow(0,0,y[0],y[1],head_width=0.1,length_includes_head=True)
plt.text(y[0],y[1],'$\mathbf{y}$')
plt.arrow(0,0,z[0],z[1],head_width=0.1,length_includes_head=True)
plt.text(z[0],z[1],'$\mathbf{z}$')

# verificando que x e y sao ortogonais
print('O produto escalar de x e y é:',np.dot(x,y))

# projetando z em x e y
projz_em_x = np.dot(z,x)*x/np.linalg.norm(x)**2
projz_em_y = np.dot(z,y)*y/np.linalg.norm(y)**2

# plotando as projeções
plt.arrow(0,0,projz_em_x[0],projz_em_x[1],head_width=0.1,length_includes_head=True)
plt.plot([z[0],projz_em_x[0]],[z[1],projz_em_x[1]],'--')

plt.arrow(0,0,projz_em_y[0],projz_em_y[1],head_width=0.1,length_includes_head=True)
plt.plot([z[0],projz_em_y[0]],[z[1],projz_em_y[1]],'--')
```

### Combinação Linear

Dados um conjunto de vetores $\mathbf{x}_1,\mathbf{x}_2,\ldots,\mathbf{x}_k$ e escalares $c_1,c_2,\ldots,c_k$, o vetor $\mathbf{Y}$ dado por

$$
\mathbf{y}=c_1\mathbf{x}_1+c_2\mathbf{x}_2+\ldots+c_k\mathbf{x}_k
$$

é uma combinação linear dos vetores $\mathbf{x}_i,\, i=1,\ldots,k$.

O conceito de combinação linear é um dos mais importantes na teoria de álgebra linear, pois permite escrever todos os vetores de um espaço vetorial como combinação de alguns poucos vetores. Por exemplo, os vetores em $\mathbb{R}^n$ podem ser escritos como combinação dos vetores $\mathbf{e}_1 = [1,0,\ldots,0]^\top,\,\mathbf{e}_2 = [0,1,\ldots,0]^\top,\cdots,\mathbf{e}_n = [0,0,\ldots,1]^\top$, ou seja,

$$
\mathbf{x}=\begin{bmatrix} x_1\\ x_2\\ \vdots\\ x_n \end{bmatrix} = 
x_1\begin{bmatrix} 1\\ 0\\ \vdots\\ 0 \end{bmatrix}+
x_2\begin{bmatrix} 0\\ 1\\ \vdots\\ 0 \end{bmatrix}+\cdots+
x_n\begin{bmatrix} 0\\ 0\\ \vdots\\ 1 \end{bmatrix}
$$

## <font color='blue'>Subespaço Vetorial</font>

Um __subespaço vetorial__ em $\mathbb{R}^n$ é um subconjunto (não vazio) $\mathcal{S}$ de $\mathbb{R}^n$ que satisfaz as 8 condições de um espaço vetorial descritas acima. Em particular, a combinação linear de quaisquer vetores em $\mathcal{S}$ é um vetor em $\mathcal{S}$.

__OBS:__ Note que o vetor nulo pertence a todos os subespaços de $\mathbb{R}^n$.

Considere, em $\mathbb{R}^3$, o plano dado pela equação $x+y-z=0$. 
Os vetores $\mathbf{x}_1=[1, 0, 1]^\top$ e $\mathbf{x}_2=[0, 1, 1]^\top$ pertencem a tal plano (como verificar esta afirmação?). Qualquer combinação linear $\mathbf{y}=a\mathbf{x}_1+b\mathbf{x}_2$ pertence ao plano, pois

$$
\mathbf{y} = \begin{bmatrix} a \\ b \\ a+b \end{bmatrix}
$$

logo o plano corresponde a um subespaço de $\mathbb{R}^3$. 

Na verdade, qualquer reta ou plano passando pela origem do espaço $\mathbb{R}^3$ corresponde a um subespaço de $\mathbb{R}^3$. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# gerando o plano x+y=z
normal = np.array([1,1,1])
xx, yy = np.meshgrid(range(-3,3), range(-3,3))
zz = normal[0] * xx + normal[1] * yy

# plot do plano
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(xx, yy, zz, alpha=0.2)

# plot da origem
ax.scatter(0,0,0,s=25,color='k')

# plot dos vetores que estao no plano z=x+y
x1 = np.array([1, 0, 1])
x2 = np.array([0, 1, 1])
ax.quiver(0, 0, 0, x1[0],x1[1],x1[2],color='blue')
ax.quiver(0, 0, 0, x2[0],x2[1],x2[2],color='red')

# plot de uma combinação linear
a = -1
b = -2
y = a*x1+b*x2
ax.quiver(0, 0, 0, y[0],y[1],y[2],color='green')


plt.show()
```

## <font color='blue'>Independência Linear, Base e Dimensão</font>

### Independência Liner

Um conjunto de vetores $\mathbf{x}_1,\ldots,\mathbf{x}_k$ em $\mathbb{R}^n$ é chamado __linearmente independente__ ($\,$LI$\,$) se a combinação linear que resulta no vetor nulo

$$
c_1\mathbf{x}_1+c_2\mathbf{x}_2+\ldots+c_k\mathbf{x}_k = 0
$$

somente é possivel quando $c_1=c_2=\cdots=c_k=0$. Em outras palavras, a única forma da combinação linear ser nula é quando todos os coeficientes são iguais a zero. 

Caso exista uma combinação linear nula com coeficientes diferente de zero, o conjunto é chamado __linearmente dependente__.

Considere o espaço $\mathbb{R}^3$, não é difícil perceber que:
- quaisquer dois vetores $\mathbf{x}_1$ e $\mathbf{x}_2$ em um subespaço dado por reta são linearmente dependentes, pois:
$$
\mathbf{x}_2 = \|\mathbf{x}_2\|\frac{\mathbf{x}_1}{\|\mathbf{x}_1\|} \rightarrow
\mathbf{x}_2 - \|\mathbf{x}_2\|\frac{\mathbf{x}_1}{\|\mathbf{x}_1\|} = 0
$$
<br>
- quaisquer três vetores $\mathbf{x}_1, \mathbf{x}_2$ e $\mathbf{x}_3$ em um subespaço dado por um plano são linearmente dependetes, pois:
$$
\mathbf{x}_3 = a\mathbf{x}_1+b\mathbf{x}_2
$$
para valores de $a$ e $b$ apropriados.
<br><br>
- quaisquer quatro vetores em  $\mathbb{R}^3$ são linearmente dependentes.

No caso particular de matrizes quadradas, isto é, matrizes com o mesmo número de linhas e colunas, se considerarmos que as colunas são vetores, então podemos verificar se os vetores colunas são LI calculando o determinante e verificando que este é diferente de zero. Ou seja:

**Propriedade** Dado um conjunto de $n$ vetores $\mathbf{x}_1,\mathbf{x}_2,\ldots,\mathbf{x}_n$ em $\mathbb{R}^n$, estão vetores serão LI se e somente se o determinante da matriz

$$
\mathbf{X} = \begin{bmatrix}
| & | & & |\\
\mathbf{x}_1 & \mathbf{x}_2 & \cdots & \mathbf{x}_k\\
| & | & & |
\end{bmatrix}
$$

for diferente de zero.

### Base de um (sub)espaço vetorial

Considere um conjunto de vetores $V =\{\mathbf{x}_1, \mathbf{x}_2,\dots,\mathbf{x}_k\}$ em um espaço $\mathbb{R}^n$. Qualquer combinação linear de tais vetores é um vetor contido no subspaço que contém os vetores de $V$, ou seja, os vetores geram um subspaço $\mathcal{S}$ de $\mathbb{R}^n$. 

**Pergunta:** é necessário todos os vetores em $V$ para gerar o subespaço $\mathcal{S}$?

_Exemplo_: Os vetores $[1,0,0],[0,1,0],[-2,-2,0]$ geram o plano $xy$ no espaço $\mathbb{R}^3$, porém, o mesmo plano é gerado somente por $[1,0,0],[0,1,0]$.

Uma **base** de um subespaço $\mathcal{S}$ é um conjunto de vetores $V$ com as seguintes propriedades:

1. Os vetores de $V$ são linearmente independentes
2. Os vetores de $V$ geram o espaço $\mathcal{S}$, ou seja, qualquer vetor em $\mathcal{S}$ é uma combinação linear dos vetores em $V$.

Desta forma, uma base do espaço $\mathcal{S}$ contém o **número mínimo de vetores** para gerar $\mathcal{S}$.

Uma consequência da definição acima é cada vetor $\mathbf{x}$ em $\mathcal{S}$ está associado a um **único conjunto de coeficientes** $c_1,\ldots,c_k$ tal que 

$$
\mathbf{x} = c_1\mathbf{x}_1+\cdots+c_k\mathbf{x}_k
$$

Isto é verdade pois se existissem outro conjunto de coeficientes tal que $\mathbf{x} = a_1\mathbf{x}_1+\cdots+a_k\mathbf{x}_k$ então teriamos que:

$$
0 = \mathbf{x} - \mathbf{x} = (c_1\mathbf{x}_1+\cdots+c_k\mathbf{x}_k) - (a_1\mathbf{x}_1+\cdots+a_k\mathbf{x}_k) = (c_1-a_1)\mathbf{x}_1+\cdots+(c_k-a_k)\mathbf{x}_k
$$

como os vetores $\mathbf{x}_i$ são LI temos que $(c_1-a_1)=\dots=(c_k-a_k)=0$, ou seja $c_i=a_i$, logo o conjunto de coeficientes é único.

Os coeficientes $c_1,\ldots,c_k$ são chamados de **coordenadas de $\mathbf{x}$ na base $V$** e cada um dos vetores $\mathbf{x}_i$ são chamados de **vetores da base**.

Existem diferentes bases para um mesmo espaço, ou seja, qualquer conjunto LI que gere o espaço é uma base.

No exemplo abaixo, os vetores $\mathbf{x}_1,\mathbf{x}_2$ e $\mathbf{x}_3$ geram o espaço $\mathbb{R}^2$, porém eles não são LI. Quais dois destes vetores formam uma base para $\mathbb{R}^2$.

```python
import numpy as np
import matplotlib.pyplot as plt

# gerando eixos de R^2
plt.plot([-1,1],[0,0],color='k')
plt.plot([0,0],[-1,1],color='k')

# plot dos vetores
x1 = np.array([0.5,0.5])
x2 = np.array([-0.5,0.7])
x3 = np.array([0.8,-0.7])
plt.arrow(0,0,x1[0],x1[1],head_width=0.1,length_includes_head=True)
plt.text(x1[0],x1[1],'$\mathbf{x}_1$')
plt.arrow(0,0,x2[0],x2[1],head_width=0.1,length_includes_head=True)
plt.text(x2[0],x2[1],'$\mathbf{x}_2$')
plt.arrow(0,0,x3[0],x3[1],head_width=0.1,length_includes_head=True)
plt.text(x3[0],x3[1],'$\mathbf{x}_3$')
```

No exemplo acima, para encontrar as coordenadas de $\mathbf{x}_3$ na base formada por $\mathbf{x}_1$ e $\mathbf{x}_2$, basta resolver o sistema gerado pela equação $\mathbf{x}_3 = c_1\mathbf{x}_1+c_2\mathbf{x}_2$:

$$
\begin{bmatrix}
0.5 & -0.5\\
0.5 & 0.7\\
\end{bmatrix}
\begin{bmatrix}
c_1\\
c_2
\end{bmatrix} = 
\begin{bmatrix}
0.8\\
-0.7
\end{bmatrix}
$$

Na equação acima, as colunas da matriz correspondem aos vetores $\mathbf{x}_1$ e $\mathbf{x}_2$.

```python
# Definindo vetores
x1 = np.array([0.5,0.5])
x2 = np.array([-0.5,0.7])
x3 = np.array([0.8,-0.7])

# fazendo os vetores x1 e x2 como colunas da matriz A
A = np.zeros((2,2))
A[:,0] = x1
A[:,1] = x2

# encontrando as coordenadas
c = np.linalg.solve(A,x3)
print('As coordendas de x3 na base {x1,x2} é:',c)

print(x3,'=',c[0]*x1+c[1]*x2)
```

### Dimensão de um (sub)espaço vetorial

Como vimos, um espaço vetorial possui uma quantidade infinita de bases, porém, todas elas possuem o mesmo número de vetores. O número de vetores na base de um espaço $\mathcal{S}$ é chamado de **dimensão do espaço**, denotado por $dim(\mathcal{S})$. 

Em particular, $\mathbb{R}^n$ tem exatamente $n$ vetores na base, logo a dimensão do espaço $\mathbb{R}^n$ é $n$. 

Um fato importante é que em qualquer espaço de dimensão $n$, nenhum conjunto com mais que $n$ vetores pode ser LI e nenhum conjunto com menos que $n$ vetores pode gerar o espaço.

## <font color='blue'> Transformação Linear</font>

Seja $T:\mathbb{R}^k\rightarrow\mathbb{R}^n$ uma transformação que mapeia vetores do espaço $\mathbb{R}^k$ para o espaço $\mathbb{R}^n$. A transformação $T$ é uma **transformação linear** se para quaisquer vetores $\mathbf{x}_1$ e $\mathbf{x}_2$

$$
T(c_1\mathbf{x}_1+c_2\mathbf{x}_2) = c_1T(\mathbf{x}_1)+c_2T(\mathbf{x}_2)
$$

onde $c_1$ e $c_2$ são números reais.

**Propriedade Importante**: Para toda transformação linear $T$ existe uma matriz $\mathbf{A}$ tal que $T(\mathbf{x})=\mathbf{A}\mathbf{x}$ e vice versa. 

A propriedade acima é verdadeira pois: considere os vetores $\mathbf{e}_1=[1,0,\ldots,0],\mathbf{e}_2=[0,1,\ldots,0],\ldots,\mathbf{e}_k=[0,0,\ldots,1]$ que formam uma base para o espaço $\mathbb{R}^k$. Qualquer vetor $\mathbf{x}=[x_1,x_2,\ldots,x_k]\in \mathbb{R}^k$ é dado pela combinação linear $\mathbf{x}=\sum_i x_i\mathbf{e}_i$. Logo:

$$
T(\mathbf{x})=T(x_1\mathbf{e}_1+\cdots+c_k\mathbf{e}_k)=x_1T(\mathbf{e}_1)+\cdots+x_kT(\mathbf{e}_k)
$$

A combinação linear acima pode ser escrita na forma matricial como:

$$
x_1T(\mathbf{e}_1)+\cdots+x_kT(\mathbf{e}_k)=
\begin{bmatrix}
| & & |\\
T(\mathbf{e}_1) & \cdots & T(\mathbf{e}_k)\\
| & & |
\end{bmatrix}
\begin{bmatrix} x_1\\ \vdots \\ x_k\end{bmatrix}
$$

Logo $T(\mathbf{x})=\mathbf{A}\mathbf{x}$, onde

$$
\mathbf{A}=
\begin{bmatrix}
| & & |\\
T(\mathbf{e}_1) & \cdots & T(\mathbf{e}_k)\\
| & & |
\end{bmatrix}
$$

## <font color='blue'> Os Quatro Espaços Fundamentais </font>

### O espaço coluna de uma matriz

Considere um conjunto de vetores $\mathbf{x}_1,\mathbf{x}_2,\ldots,\mathbf{x}_k$ em $\mathbb{R}^n$. Tais vetores podem ser organizados como colunas de uma matriz $\mathbf{X}$:

$$
\mathbf{X} = \begin{bmatrix}
| & | & & |\\
\mathbf{x}_1 & \mathbf{x}_2 & \cdots & \mathbf{x}_k\\
| & | & & |
\end{bmatrix}
$$

O espaço gerado pelos vetores $\mathbf{x}_1,\mathbf{x}_2,\ldots,\mathbf{x}_k$ é chamado de **espaço coluna** da matriz $\mathbf{X}$ e é denotado por $\mathcal{C}(\mathbf{X})$. A dimensão do espaço coluna é chamado de **posto da matriz** e seu valor $r\leq k$.

**OBS**: O espaço coluna também é chamado de **espaço imagem** da transformação linear representada pela matriz.

### O espaço nulo de uma matriz

Dada uma matriz $\mathbf{X}$, $n\times k$ ($n$ linhas e $k$ colunas) e um vetor $\mathbf{c}\in\mathbb{R}^k$, a multiplicação $\mathbf{X}\mathbf{c}$ corresponde  a uma combinação linear das colunas de $\mathbf{X}$ com coeficientes dados pelas entradas de $\mathbf{c}$, ou seja:

$$
\mathbf{X}\mathbf{c} = 
\begin{bmatrix}
| &  & |\\
\mathbf{x}_1  & \cdots & \mathbf{x}_k\\
| &  & |
\end{bmatrix}
\begin{bmatrix}
c_1 \\ \vdots \\ c_k
\end{bmatrix}=
c_1\mathbf{x}_1 + \cdots + c_k\mathbf{x}_k
$$

O **espaço nulo** da matriz $\mathbf{X}$, denotado $\mathcal{N}(\mathbf{X})$, corresponde ao espaço dos vetores $\mathbf{c}$ tal que

$$
\mathbf{X}\mathbf{c} = \mathbf{0}
$$

**OBS**: O espaço nulo é também é chamado de **núcleo da transformação linear** representada pela matriz.

### O espaço linha de uma matriz

O **espaço linha** de uma matriz $\mathbf{X}$ é o espaço gerado pelos vetores que correspondem às linhas na matriz, que é equivalente ao espaço coluna da matriz $\mathbf{X}^\top$, ou seja, $\mathcal{C}(\mathbf{X}^\top)$.

### O espaço nulo à esquerda de uma matriz

O **espaço nulo à esquerda** de $\mathbf{X}$ é o espaço formado pelos vetores $\mathbf{c}$ tal que 

$$
\mathbf{c}^\top\mathbf{X}=\mathbf{0}
$$

Note que a equação acima pode ser escrita como 

$$
\mathbf{X}^\top\mathbf{c}=\mathbf{0}
$$

Desta forma, espaço nulo à esquerda é o espaço nulo de $\mathbf{X}^\top$, ou seja, $\mathcal{N}(\mathbf{X}^\top)$ 

<font color='red'>Fatos importantes:</font>

1. A matriz $\mathbf{X}$ é uma transformação que leva vetores do espaço $\mathbb{R}^k$ (domínio da transformação) para o espaço $\mathbb{R}^n$ (imagem da transformação), ou seja:

$$
\mathbf{X}: \mathbb{R}^k \rightarrow  \mathbb{R}^n
$$

2. A matriz transposta dá origem a uma transformação no sentido contrário, ou seja:

$$
\mathbf{X}^\top: \mathbb{R}^n \rightarrow  \mathbb{R}^k
$$

$\qquad\qquad\qquad\qquad\qquad\qquad$(cuidado para não confundir com a transformação inversa)

3. A dimensão  do espaço nulo é diferente de zero somente se as coluna de $\mathbf{X}$ são linearmente dependentes
<br>

4. Os espaços $\mathcal{C}(\mathbf{X})$ e $\mathcal{N}(\mathbf{X}^\top)$ são subespaços contidos em $\mathbb{R}^n$
<br>

5. Os espaços $\mathcal{C}(\mathbf{X}^\top)$ e $\mathcal{N}(\mathbf{X})$ são subespaços contidos em $\mathbb{R}^k$

## <font color='blue'> Teorema do Núcleo e da Imagem </font>

**Propriedade:** A dimensão do espaço coluna de uma matriz $\mathbf{X}$ é igual a dimensão do espaço linha de $\mathbf{X}$.
<br>

Veja a referência [https://www.ime.usp.br/~oliveira/POSTOMACKIW.pdf](https://www.ime.usp.br/~oliveira/POSTOMACKIW.pdf) para uma demostração deste fato.

**Teorema do Núcleo e da Imagem**: Dada uma matriz $\mathbf{X}$ com $n$ linhas e $k$ colunas,

$$
dim(\mathcal{N}(\mathbf{X}))+dim(\mathcal{C}(\mathbf{X})) = k
$$

Uma consequência do teorema acima é que se a $dim(\mathcal{C}(\mathbf{X}))=r$, então:

1. $dim(\mathcal{C}(\mathbf{X}^\top))=r$
2. $dim(\mathcal{N}(\mathbf{X})) = k-r$
3. $dim(\mathcal{N}(\mathbf{X}^\top)) = n-r$
